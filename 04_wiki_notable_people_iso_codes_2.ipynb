## Impute selected ISO3 values into non-ISO3 values for Plotting-purposes only

#### Citizenship = "Austria-Hungary"

```python
iso_3["map_iso3"] = iso_3["map_iso3"].str.replace("\['AUT', 'HUN'\]", "AUT")
```

```python
iso_3.loc[iso_3["citizenship"] == "Austria-Hungary"].head()
```

```python
iso_3.loc[iso_3["citizenship"] == "Roman Empire"].head()
```
